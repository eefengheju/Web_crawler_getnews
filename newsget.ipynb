```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

# 配置 Chrome 浏览器选项
chrome_options = Options()
chrome_options.add_argument('--headless')  # 开启 headless 模式

# 创建一个 Chrome 浏览器对象
driver = webdriver.Chrome(options=chrome_options)

# 访问网页
url ='http://www.china.com.cn/app/template/amucsite/web/webLive.html#3032'
driver.get(url)

# 等待页面加载完毕
driver.implicitly_wait(10)

# 获取页面内容
html = driver.page_source

# 解析页面内容
soup = BeautifulSoup(html, 'html.parser')

# 关闭浏览器
driver.quit()
```

找到发言内容的位置

```python
div_list = soup.find_all('div', {'class': 'txt'})
del div_list[-1]
news_text=''
for p in div_list :
    news_text += p.text.strip()+ '\n'
print(news_text)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
dataf = pd.read_excel('data4.xlsx')  #设置成你的xlsx表格名称
title = dataf['标题'] #与xlsx表格中表头的信息一致
channelName = []
net = dataf['链接']
content = []
print(dataf.head())
```

```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.chrome.options import Options
chrome_options = Options()

chrome_options.add_argument('--headless')  # 开启 headless 模式
driver = webdriver.Chrome(options=chrome_options)


num = 0
for link_href in net:
    try:
        driver = webdriver.Chrome(options=chrome_options)
        # 访问网页
        driver.get(link_href)
        # 等待页面加载完毕
        driver.implicitly_wait(10)
        # 获取页面内容
        html = driver.page_source
        # 解析页面内容
        soup = BeautifulSoup(html, 'html.parser')
        # 关闭浏览器
        div_list = soup.find_all('div', {'class': 'txt'})
        del div_list[-1]
        news_text=''
        for p in div_list :
            news_text += p.text.strip()+ '\n'   
        content.append(news_text)
        num += 1
        print(num)
        driver.quit()
    except:
        pass
title = title[:len(content)]
#使用dataframe来进行存储
total_data = {'content': content, 'url': link_href, 'title': title}
```

保存到xlsx中

```python
data2 = pd.DataFrame(total_data)
#保存成EXCEL格式
data2.to_excel('新闻.xlsx')

```
